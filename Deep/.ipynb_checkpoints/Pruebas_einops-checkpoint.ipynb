```python
import einops
import numpy
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import PIL
import os
```

Se importa un conjunto de datos.

```python
imagenes = []
for filepath in os.listdir('imagenes_ejemplo/'):
    imagenes.append(numpy.array(PIL.Image.open('imagenes_ejemplo/'+filepath)))
plt.imshow(imagenes[0])
print(imagenes[0].shape)
```

```python
def dibuja_imagen(imagen):
    plt.imshow(imagen)
```

```python
# apilar la lista de imágenes para crear un batch:
image_batch = einops.rearrange(imagenes, 'b h w c -> b h w c')
print('Forma del batch:', image_batch.shape)
# cortar la imagen en cuadritos más pequeños:
cuadritos = einops.rearrange(image_batch, 'b (p1 h) (p2 w) c -> b (p1 p2) h w c', p1=4, p2=4)
print('Forma del batch cortado en cuadritos:', cuadritos.shape)
# pegar los cuadritos en una fila:
patch = einops.rearrange(cuadritos, 'b1 b2 h w c -> b1 h (b2 w) c')
print('Forma del batch reorganizado:', patch.shape)
for i in range(patch.shape[0]):
    dibuja_imagen(patch[i])
```
